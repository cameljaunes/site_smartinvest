```python
import pandas as pd
import sqlite3

# Charger le fichier CSV dans un DataFrame pandas
chemin_csv = "C://Users//Lina//Documents//Projet_integratif//infos_table_agence.csv"
df = pd.read_csv(chemin_csv, encoding="UTF-8", delimiter=';')

# Se connecter à la base de données SQLite
connection_sqlite = sqlite3.connect("C://Users//Lina//Documents//sqlite (5).db")
curseur = connection_sqlite.cursor()

# Correspondance entre les colonnes du CSV et les colonnes de chaque table
correspondances = {
    'Agence': {'CodeAG': 'CodeAG', 'NomAG': 'NomAG', 'AdresseAG': 'AdresseAG', 'TelAG': 'TelAG', 'MailAG': 'MailAG', 'CodeQ' : 'CodeQ'}
}

# Parcourir chaque ligne du DataFrame et insérer les données dans les tables correspondantes
for index, row in df.iterrows():
    try:
        for table, correspondance in correspondances.items():
            colonnes_table = ', '.join(colonne for colonne in correspondance.values())
            valeurs = [row[colonne_csv] for colonne_csv in correspondance.keys()]
            curseur.execute(f"INSERT INTO {table} ({colonnes_table}) VALUES ({', '.join(['?']*len(valeurs))})", tuple(valeurs))

        print(f"Insertion de la ligne {index} réussie")
    except Exception as e:
        print(f"Erreur lors de l'insertion de la ligne {index}: {e}")

# Valider les changements et fermer la connexion
connection_sqlite.commit()
connection_sqlite.close()
```

```python
import csv
from faker import Faker

fake = Faker()

# Generating 50 random name and surname pairs
random_data = [(fake.first_name(), fake.last_name()) for _ in range(50)]

# Writing the data to a CSV file
csv_file_path = "C://Users//Lina//Documents//random_names.csv"
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Surname"])  # Writing header
    writer.writerows(random_data)  # Writing the random name and surname pairs

print(f"CSV file '{csv_file_path}' created successfully.")
```
